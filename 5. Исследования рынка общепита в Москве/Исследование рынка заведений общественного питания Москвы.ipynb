## Изучение и предобработка данных

```python
import pandas as pd
from plotly import graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

rest_data = pd.read_csv('/datasets/rest_data.csv')
rest_data.head(5) 
```

```python
rest_data.tail(5) 
```

```python
rest_data.sample(5) 
```

**Таблица `rest_data` содржит 6 столбцов**

- `id` — идентификатор объекта;

- `object_name` — название объекта общественного питания;

- `chain` — сетевой ресторан;

- `object_type` — тип объекта общественного питания;

- `address` — адрес;

- `number` — количество посадочных мест.

```python
rest_data.info()
```

Пропусков нет, тип данных - корректен.

```python
rest_data.describe()
```

Проверим позиции в таблице, где количество посадочных мест равно 0

```python
print('Кол-во заведений без посадочных мест:', len(rest_data.query('number==0')))
rest_data[rest_data['number'] == 0]['object_name'].to_list()
```

Скорее всего данные заведения работают "на доставку"/"на вынос", так же при заполнении данных могли не указать кол-во посадочных мест.

Проверим строки на дубликаты

```python
print('Общее кол-во дубликатов:',rest_data.duplicated().sum())
print('Кол-во дубликатов без id:',len(rest_data[rest_data[['object_name', 'object_type', 'address']].duplicated()]))
rest_data[rest_data[['object_name', 'object_type', 'address']].duplicated()].head(15)
```

В таблице есть дубликаты по наименованию и адресу, но они так же могут отличаться по типу заведения и количеству посадочных мест, а так же `id`. Возможно данные по заведениям вносились заново и соответственно присваивался новый `id`. Поскольку это никак нельзя уточнить, а количество таких дубликатов ~1% от общего количества данных, что не критично, оставим данные дубликаты без изменения.

Просмотрим есть ли какие-то нетипичные значения в столбцах `chain` и `object_type`

```python
rest_data['chain'].value_counts()
```

```python
rest_data['object_type'].value_counts()
```

Заменим некоторые категории с длинными названиями, что бы график смотрелся лаконично. Так же для удобства изменим значения `chain`.

```python
rest_data['object_type'] = rest_data['object_type'].str.replace('магазин \(отдел кулинарии\)', 'кулинария') 
rest_data['object_type'] = rest_data['object_type'].str.replace('предприятие быстрого обслуживания', 'пбо')
```

```python
rest_data['chain'] = rest_data['chain'].str.replace('да', 'Сетевое') 
rest_data['chain'] = rest_data['chain'].str.replace('нет', 'Несетевое')
```

**Вывод**
- Тип данных корректен для каждого столбца;
- Пропуски в таблице отсутствуют;
- Количество дубликатов некритичное для анализа (~1%), оставлены без изменений;
- Просмотрены заведения без посадочных мест. Данные заведения, скорее всего работают "на вынос"/"на доставку" или 0 заполняли пропуски в таблице.

## Анализ данных

### Исследуем соотношение видов объектов общественного питания по количеству.

```python
object_type = rest_data.groupby('object_type').agg({'id':'count'}).rename(columns={'id':'counts'})
object_type['percent'] = object_type['counts'] / object_type.counts.sum()*100
object_type.columns = ['Всего', '%']
object_type.sort_values(by='%', ascending=False)
```

```python
object_type = rest_data.pivot_table(index='object_type', values='id', aggfunc='count').sort_values(by='id', ascending=False).reset_index()

plt.figure(figsize=(10, 7))
sns.barplot(x='object_type', y='id', data=object_type, palette='Blues_d')
plt.title('Соотношение типов заведений и их количества', fontsize=20) 
plt.xlabel('Тип заведения')
plt.ylabel('Количество заведений')
plt.show()
```

**Вывод**

- С большим отрывом лидирует кафе - 39,7% от общего числа заведений (6099 точек);
- Второе и третье место занимает столовая - 16,8% (2587) и рестораны - 14,%(2285);
- Наименее популярный тип заведений кулинария - 1,7% (всего 273 заведения).

### Исследуем соотношение сетевых и несетевых заведений по количеству.

```python
chain = rest_data.groupby('chain').agg({'id':'count'}).rename(columns={'id':'counts'})
chain['percent'] = chain['counts'] / chain.counts.sum()*100
chain.columns = ['Всего', '%']
chain.sort_values(by='%', ascending=False)
```

```python
plt.figure(figsize=(7, 7))
colors = ('tab:blue','tab:red') 
rest_data['chain'].value_counts().plot.pie(colors = colors, label = '', labels = ['Несетевые заведения', 'Сетевые заведения'], autopct = '%1.1f%%')
#sns.barplot(x='chain', y='id', data=chain, palette='Blues_d')
plt.title('Соотношение сетевых и несетевых заведений', fontsize=20) 
plt.show()
```

**Вывод**

- Большая часть заведений в Москве несетевые - 80,6% (12 398 заведений);
- Сетевых заведений в 4 раза меньше, всего 19,4% (2 968 заведений).

### Для какого вида объекта общественного питания характерно сетевое распространение?

```python
type_rest = rest_data.pivot_table(index='object_type',columns='chain', values='id', aggfunc='count').reset_index()
type_rest['percent'] = (type_rest['Сетевое'] /(type_rest['Сетевое']+type_rest['Несетевое']))*100
type_rest.columns = ['Тип объекта','Несетевое', 'Сетевое', '% Сетевых']
type_rest.sort_values(by='% Сетевых', ascending=False)
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='object_type', data=rest_data, hue='chain', palette='Set1') 
plt.title('Соотношение сетевых и несетевых объектов общественного питания по видам', fontsize=20)
plt.xlabel('Вид объекта')
plt.ylabel('Количество заведений')
plt.show()
```

**Вывод**

- Сетевое распространения характерно для предприятий быстрого обслуживания - 41,1% от общего числа данного типа заведений.

- Так же довольно часто встречаются сетевые заведения среди кулинарии - 28,5%, ресторанов - 22,9%, кафе - 22,8%.

### Что характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?

Отфильтруем данные:
- Поскольку вопрос подразумевает какое-то наличие посадочных мест уберем из выборки заведение, где их 0;
- Выберем только сетевые заведения. 

```python
rest_data_filtered = rest_data[rest_data['number']>0] 
rest_data_filtered = rest_data[rest_data['chain'] == 'Сетевое']
```

```python
rest_data_filtered = rest_data_filtered.groupby('object_name').agg({'id':'count', 'number':'mean'})

plt.figure(figsize = (15, 7))
#sns.distplot(rest_data_filtered['number'], bins=100)
plt.scatter(x = 'number', y = 'id', data=rest_data_filtered, alpha=0.2)
plt.title('Соотношение количества заведений сети к количеству посадочных мест', fontsize=20) 
plt.xlabel('Количество посадочных мест')
plt.ylabel('Количество заведений в сети')
plt.show()
```

**Вывод** 

На графике видно, что наиболее плотное скопление сосредоточено в отрезке до 50 . Значит, что для сетевых заведений характерно иметь больше точек с небольшим количеством посадочных мест.

### Для каждого вида объекта общественного питания опишем среднее количество посадочных мест. Какой вид предоставляет в среднем самое большое количество посадочных мест?

```python
number = rest_data.groupby('object_type').agg({'id':'count','number':'mean'}).rename(columns={'id':'counts'})
number.columns = ['Всего', 'Среднее']
number.sort_values(by='Среднее', ascending=False)
```

```python
number = rest_data.groupby('object_type').agg({'number':'mean'}).reset_index().sort_values(by='number', ascending=False)

plt.figure(figsize=(10, 7))
ax = sns.barplot(x='object_type', y='number', data=number, palette='Blues_d')
plt.title('Среднее количество посадочных мест по типу заведения', fontsize=20)
plt.xlabel('Тип заведения')
plt.ylabel('Количество посадочных мест')
plt.show()
```

**Вывод**

- Больше всего посадочных мест в столовых - 130 (средний показатель);
- Наименьший показатель в магазинах (отдел кулинарии), всего 5 посадочных мест.

### Построим график топ-10 улиц по количеству объектов общественного питания. В каких районах Москвы находятся эти улицы?

Выделим в отдельный столбец информацию об улице из столбца `address` .

```python
rest_data['street'] = rest_data['address'].str.split(',', expand=True)[1]
rest_data['street'] = rest_data['street'].str.strip()
rest_data.head()
```

```python
top_streets = rest_data.groupby('street').agg({'id':'count'}).reset_index().sort_values(by='id', ascending=False)
top_streets = top_streets.head(10)
top_streets.reset_index(drop=True, inplace=True)
top_streets.columns = ['Улица', 'Всего']
top_streets
```

Столбец с адресом имел структурированный вид, но после группировки наиболее распространённых улиц встречаются адреса в Зеленограде и поселении Сосенское, исключим эти записи.

```python
top_streets_gr = rest_data.query("street != 'город Зеленоград' and street != 'поселение Сосенское'").groupby('street').agg({'id':'count'}).sort_values(by='id', ascending=False).reset_index().head(10)

plt.figure(figsize=(10, 7))
sns.barplot(x='id', y='street' , data=top_streets_gr, palette='Blues_d');
plt.title('Топ 10 улиц с наибольшим количеством заведений общественного питания', fontsize=15) 
plt.xlabel('Количество заведений')
plt.ylabel('Название улицы') 
plt.show()
```

**Вывод**

- Медианное количество посадочных мест - 40;

•	Больше всего заведений располагаются на: проспекте Мира (204 точки), Профсоюзной улице (182 точки) и Ленинградском проспекте (171 точка);

•	На территории Москвы находится 622 улицы с 1 объектом общественного питания. Больше всего таких улиц в ЦАО, а конкретно в Таганском районе (27 улиц);

•	Медианное количество посадочных мест - 40. Вне зависимости от улицы и типа заведения общее число мест не превышает - 100.

## Рекомендации

**Вид заведения**

Открыть именно робот-кафе - хорошее решение, т.к. этот тип заведений наиболее популярен и привычен среди потребителей. Преимуществом можно считать небольшое и простое меню, а также спрос у посетителей разных возрастных категорий, что не скажешь, допустим, о ресторанах и барах. 

Решение об открытии сети робот-кафе стоит принять, после проведения анализа работы данного заведения: его рентабельности, спроса, среднего чека, хотя бы в течении года после запуска. Сейчас нет никаких оснований планировать сеть. Тем не менее на рынке заведений общественного питания Москвы преобладают сети с небольшим количеством точек питания, так что в дальнейшем, после сбора информации, можно задуматься о развитии в этом направлении.

**Количество посадочных мест**

Наиболее распространенное число посадочных мест в кафе Москвы от 30 до 50. Поскольку формат заведения подразумевает недолгий перекус, то небольшое количество посадочных мест вполне оправданно. 

Опять-таки сравнивая с рестораном, в котором более длительно время провождения, соответственно необходимо больше мест для размещения гостей. Так же небольшая вместительность заведения, и соответственно площадь, возможно, поможет сэкономить на аренде.

**Место расположения**

Можно рекомендовать разместить заведение в ЦАО, в данном округе много улиц с одним заведением, что позволит не потеряться среди конкурентов, плюсом так же является высокая проходимость. 

Вряд ли можно рекомендовать конкретную улицу. Необходимо её определить исходя из параметров: стоимости аренды, расположения близлежащих объектов, рассчитать проходимост и т.д.
